```python
import matplotlib.pyplot as plt

# problem sizes
problem_sizes = [1000, 10000, 100000]

# Runtimes in seconds
runtimes_cpu = [0.019, 0.960, 286.102]  # Sequential CPU runtimes
runtimes_cuda_loop = [0.377, 0.976, 71.189]  # CUDA loop-based implementation runtimes
runtimes_cuda_task = [0.467, 1.025, 59.512]  # CUDA task-based implementation runtimes

# Plotting
plt.figure(figsize=(10, 6))
plt.scatter(problem_sizes, runtimes_cpu, label='Sequential (CPU)', marker='o')
plt.scatter(problem_sizes, runtimes_cuda_loop, label='CUDA Loop', marker='o')
plt.scatter(problem_sizes, runtimes_cuda_task, label='CUDA Task', marker='o')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Problem Size (N)')
plt.ylabel('Runtime (seconds)')
# plt.title('Runtime Comparison of Number Crunching Implementations')
plt.legend()
# plt.grid(True)

plt.show()
```
