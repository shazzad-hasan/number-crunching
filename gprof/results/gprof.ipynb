```python
# gprof result
import matplotlib.pyplot as plt

# Given N values and corresponding execution times
N_values = [k * 10000 for k in range(1, 11)]  # N from 10,000 to 100,000
execution_times = [0.17, 0.69, 1.56, 2.79, 4.98, 8.08, 4.97, 7.85, 12.75, 10.33]  # seconds

plt.figure(figsize=(10, 6))
plt.plot(N_values, execution_times, marker='o', linestyle='-', color='b')
# plt.title('Execution Time vs Problem Size N')
plt.xlabel('Problem Size N')
plt.ylabel('Execution Time (seconds)')
plt.grid(True)
plt.show()
```
